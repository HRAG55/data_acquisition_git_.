# 1️⃣ Dataset Selection

**Dataset:** Titanic Dataset  
**Source:** [Data Science Dojo GitHub](https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv)  

**Data Collection Methodology:**  
- The dataset comes from the Titanic ship voyage in 1912, containing passenger information such as age, sex, class, and survival.  
- The dataset is publicly available on Kaggle and GitHub, accessible without restrictions.  

**Reliability & Limitations:**  
- It is a historical dataset widely used in machine learning, so it is reliable for educational and analytical purposes.  
- Some values are missing (like Age or Cabin), which may affect analysis if not handled properly.  
- The data shows clear biases, such as females and upper-class passengers having higher survival rates. This should be considered when building models.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
df.head()
```

```python
print("Shape:", df.shape)

print(df.dtypes)

print(df.describe())

print(df.isnull().sum())
```

```python
# 1️⃣ Survival Count by Gender
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title("Survival Count by Gender")
plt.show()

# 2️⃣ Age Distribution of passengers
sns.histplot(df['Age'].dropna(), bins=30)
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

## Insights
- Females survived significantly more than males.  
- Most passengers were between 20 and 40 years old.


## Ethical & Legal Reflection

**License & Privacy:**  
- The dataset is public and available for everyone from Kaggle.  
- It does not contain modern personal data, so GDPR does not directly apply.

**Reflection:**  
The Titanic dataset is historical and publicly available, so privacy issues are minimal. However, ethical handling still matters. Missing data, especially age and cabin information, can bias models if not handled carefully. Gender and class biases are evident in survival rates, demonstrating the importance of considering fairness in AI models. When building predictive models, ensuring transparency about dataset limitations, acknowledging biases, and not overgeneralizing conclusions is crucial. Even with historical datasets, responsible data use involves ethical reflections on bias, fairness, and reliability of insights drawn.
